```python
from bs4 import BeautifulSoup as bs
import requests
import pymongo
from splinter import Browser
import pandas as pd
import time
```

```python
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019]

player_stats=[]
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

for i in range(len(year)):
    url=f"https://www.pro-football-reference.com/years/{year[i]}/receiving.htm"
    browser.visit(url)
    time.sleep(4)
    page_html=browser.html
    soup = bs(page_html, 'lxml')
    tables=soup.find_all('tbody')
    players=tables[10]
    recievers=players.find_all('tr')
    for x in range(len(recievers)):
        try:
            r=recievers[x].th.text
            ranks=r
            name=recievers[x].td.a.text
            t=recievers[x]('td', attrs={'data-stat': 'team'})
            te=t[0]('a')
            team=te[0].text
            a=recievers[x]('td', attrs={'data-stat': 'age'})
            ag=a[0].text        
            games_text=recievers[x]('td', attrs={'data-stat': 'gs'})
            g=games_text[0].text
            targets=recievers[x]('td', attrs={'data-stat': 'targets'})
            targets=targets[0].text
            rec=recievers[x]('td', attrs={'data-stat': 'rec'})
            rec=rec[0].text
            yards_text=recievers[x]('td', attrs={'data-stat': 'rec_yds'})
            yards=yards_text[0].text
            tds_text=recievers[x]('td', attrs={'data-stat': 'rec_td'})
            tds=tds_text[0].text
            st_text=recievers[x]('td', attrs={'data-stat': 'rec_first_down'})
            rec_first_down=st_text[0].text
            fumble_text=recievers[x]('td', attrs={'data-stat': 'fumbles'})
            fumbles=fumble_text[0].text
            player_dict={
                "rank":int(ranks),
                "name":name,
                "teams":team,
                "ages":int(ag),
                "games":int(g),
                "targets":int(targets),
                'rec':int(rec),
                "yards":int(yards),
                "touchdowns":int(tds),
                "first_downs":int(rec_first_down),
                "drops":int(fumbles),
                'season':year[i]
            }
            player_stats.append(player_dict)
        except:
#             rushers.pop(x)
            print('************skipped thead ************')
```
