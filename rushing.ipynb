```python
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019]

player_stats=[]
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

for i in range(len(year)):
    url=f"https://www.pro-football-reference.com/years/{year[i]}/rushing.htm"
    browser.visit(url)
    time.sleep(4)
    page_html=browser.html
    soup = bs(page_html, 'lxml')
    tables=soup.find_all('tbody')
    players=tables[10]
    rushers=players.find_all('tr')
    for x in range(len(rushers)):
        try:
            r=rushers[x].th.text
            ranks=r
            name=rushers[x].td.a.text
            t=rushers[x]('td', attrs={'data-stat': 'team'})
            te=t[0]('a')
            team=te[0].text
            a=rushers[x]('td', attrs={'data-stat': 'age'})
            ag=a[0].text        
            games_text=rushers[x]('td', attrs={'data-stat': 'gs'})
            g=games_text[0].text
            att_text=rushers[x]('td', attrs={'data-stat': 'rush_att'})
            rush_att=att_text[0].text
            yards_text=rushers[x]('td', attrs={'data-stat': 'rush_yds'})
            rush_yds=yards_text[0].text
            tds_text=rushers[x]('td', attrs={'data-stat': 'rush_td'})
            tds=tds_text[0].text
            st_text=rushers[x]('td', attrs={'data-stat': 'rush_first_down'})
            first_down_rush=st_text[0].text
            fumble_text=rushers[x]('td', attrs={'data-stat': 'fumbles'})
            fumbles=fumble_text[0].text
            player_dict={
                "rank":int(ranks),
                "name":name,
                "teams":team,
                "ages":int(ag),
                "games":int(g),
                "atts":int(rush_att),
                "yards":int(rush_yds),
                "touchdowns":int(tds),
                "first_downs":int(first_down_rush),
                "drops":int(fumbles),
                'season':year[i]
            }
            player_stats.append(player_dict)
        except:
#             rushers.pop(x)
            print('************skipped thead ************')
```

```python
player_df=pd.DataFrame(player_stats)
player_df.to_csv('rushing_stats.csv')
```
