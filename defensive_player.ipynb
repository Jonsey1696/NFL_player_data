```python
year=[2011,2012,2013,2014,2015,2016,2017,2018,2019]

player_stats=[]
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

for i in range(len(year)):
    url=f"https://www.pro-football-reference.com/years/{year[i]}/defense.htm"
    browser.visit(url)
    time.sleep(4)
    page_html=browser.html
    soup = bs(page_html, 'lxml')
    tables=soup.find_all('tbody')
    players=tables[10]
    defenders=players.find_all('tr')
    for x in range(len(defenders)):
        try:
            r=defenders[x].th.text
            ranks=int(r)
            name=defenders[x].td.a.text
            t=defenders[x]('td', attrs={'data-stat': 'team'})
            te=t[0]('a')
            team=te[0].text
            a=defenders[x]('td', attrs={'data-stat': 'age'})
            ag=a[0].text 
            games_text=defenders[x]('td', attrs={'data-stat': 'gs'})
            g=games_text[0].text
            inte=defenders[x]('td', attrs={'data-stat': 'def_int'})
            inte=inte[0].text
            p_defend=defenders[x]('td', attrs={'data-stat': 'pass_defended'})
            p_defend=p_defend[0].text
            fumble=defenders[x]('td', attrs={'data-stat': 'fumbles_forced'})
            fumble=fumble[0].text
            recovered=defenders[x]('td', attrs={'data-stat': 'fumbles_rec'})
            recovered=recovered[0].text
            sacs=defenders[x]('td', attrs={'data-stat': 'sacks'})
            sacs=sacs[0].text
            tackles=defenders[x]('td', attrs={'data-stat': 'tackles_combined'})
            tackles=tackles[0].text
            tackles_loss=defenders[x]('td', attrs={'data-stat': 'tackles_loss'})
            tackles_loss=tackles_loss[0].text
            qb_hits=defenders[x]('td', attrs={'data-stat': 'qb_hits'})
            qb_hits=qb_hits[0].text
            player_dict={
                "rank":int(ranks),
                "name":name,
                "teams":team,
                "ages":int(ag),
                "games":int(g),
                "interceptions":inte,
                'p_defend':p_defend,
                "forced_fumbles":fumble,
                "rec_fumbles":recovered,
                "sacks":sacs,
                "tackles":tackles,
                'tackles_loss':tackles_loss,
                'qb_hits':qb_hits,
                'season':year[i]
            }
            player_stats.append(player_dict)
        except:
            print('************skipped thead ************')
```

```python
player_df=pd.DataFrame(player_stats)
player_df.to_csv('defensive_player_stats.csv')
```
